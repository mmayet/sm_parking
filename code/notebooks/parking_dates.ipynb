```python
df.dt_hour = pd.to_datetime(df.dt_hour)
```

```python
def plot_year_garage(year, garage, time=12):
    plt.figure(figsize=(20,10))
    hour_mask = df.dt_hour.dt.hour==time
    year_mark = df.dt_hour.dt.year==year
    garage_mask = df.lot == garage
    plt.plot(df[year_mark & hour_mask & garage_mask]['dt_hour'].dt.date, 
             df[year_mark & hour_mask & garage_mask]['hr_avail'])
```

```python
plot_year_garage(2015, 'Structure 3')
```

```python
def plot_expo_diff(garage, time):
    year_before = (df.dt_hour >= '2015-04-20') & (df.dt_hour < '2016-04-20')
    year_after = (df.dt_hour >= '2016-04-20') & (df.dt_hour < '2017-04-20')
    hour_mask = df.dt_hour.dt.hour==time
    garage_mask = df.lot == garage
    
    plt.figure(figsize=(20,10))
    plt.plot(['M', 'T', 'W', 'Th', 'F', 'Sa', 'Su'], 
             df[year_before & hour_mask & garage_mask].groupby(df.dt_hour.dt.weekday)['hr_avail'].mean())
    plt.plot(['M', 'T', 'W', 'Th', 'F', 'Sa', 'Su'], 
             df[year_after & hour_mask & garage_mask].groupby(df.dt_hour.dt.weekday)['hr_avail'].mean())
    plt.legend([2015, 2016])
```

```python
plot_expo_diff('Structure 4', 14)
```

```python
def plot_hourly_average(structure, year):
    garage_mask = df.lot == f'Structure {structure}'
    year_mask = df.dt_hour.dt.year==year
    
    cmap = plt.cm.RdGy
    cmaplist = [cmap(i) for i in range(cmap.N)]
    
    plt.figure(figsize=(20,10))
    bar_list = plt.bar(range(168), df[garage_mask & year_mask].groupby(['weekday', 'hour'])['hr_avail'].mean())
    plt.xticks(range(0, 7*24, 24), ['M', 'T', 'W', 'Th', 'F', 'Sa', 'Sun'])
    plt.title(f'Hourly Average by Weekday for Structure {structure}, {year}')
    for i in range(24):
        for j in range(7):
            bar_list[i + j*24].set_color(cmaplist[(128+10*i)%256]) 
    
```

```python
plot_hourly_average(8, 2017)
```
